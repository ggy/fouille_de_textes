```julia
using Word2Vec
using Gadfly 
```

```julia
word2vec("Auteurs.txt", "Auteurs-vec.txt", verbose=true)
word2phrase("Auteurs.txt", "Auteursphrase")
word2clusters("Auteurs.txt", "Auteurs-class.txt", 100)
```

```julia
model = wordvectors("Auteurs-vec.txt")
```

```julia
words = vocabulary(model)
```

```julia
idxs, dists = cosine(model, "femme", 30)
```

```julia
plot(x=words[idxs], y=dists)
```

```julia
indxs, dists = analogy(model, ["guerre", "homme"], ["femme"], 30)
```

```julia
plot(x=words[indxs], y=dists)
```
